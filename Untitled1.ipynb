```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
X = np.array(sorted(list(range(5))) * 20) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))) * 20) + np.random.normal(size=100, scale=0.25)

print(X)
print(y)
```

```python
lr = 0.001
epho=2000
a=0 # 기울기
b=0 # 절편

for i in range(epho):
  y_pred = a*X+b #y의 예측값을 구함
  error = y-y_pred # 오차 error 값 y값 - y예측 값

  loss = 0.5*np.mean((error)**2)

  # print(loss)
  
  a_diff = -(1/len(X))*sum(X*(error))
  b_diff = -(1/len(X))*sum(y-y_pred) # 
  a = a-lr*a_diff
  b= b-lr*b_diff
    
  if i % 100==0:
    print("loss = %f ,epoch= %f,기울기=%.04f,절편=%.04f,에러=%.04f"%(loss,i,a,b,error.mean()))
    
plt.scatter(X,y)
plt.plot([min(X),max(X)],[min(y),max(y)])
plt.show()
```
